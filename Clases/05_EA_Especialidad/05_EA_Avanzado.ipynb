# Computación Evolutiva

> Metaheurísticas **bioinspiradas** basadas en los principios de la **Teoría de la evolución y la selección natural** de Charles Darwin.

Una **metaheurística** es una estrategia de alto nivel, independiente del problema, que guía la búsqueda de soluciones aproximadas de buena calidad en espacios donde una búsqueda exhaustiva es inviable.

* Gran auge desde los 90s.
* Creadas en los 70s.
* Llamados **Algoritmos Evolutivos** (_EA, Evolutionary Algorithms_).
* Pioneras en algoritmos basados en la biología y la naturaleza.
* Amplían la búsqueda a varios candidatos simultáneos (una **población**).
* Se perturban los candidatos con operadores evolutivos clásicos (selección, cruza y mutación), entre otros.

## Algoritmos Evolutivos: Framework

Todo EA comparte un mismo ciclo iterativo: se **inicializa** una población de soluciones candidatas, se **evalúa** su aptitud (_fitness_), se **seleccionan** padres, se generan hijos mediante **recombinación** y **mutación**, y se decide qué individuos **sobreviven** a la siguiente generación. El proceso se repite hasta cumplir un criterio de término.

* Holland (1973), rectificado por De Jong (1975).
* Algoritmo básico y pionero de la Computación Evolutiva.
* Creación de generaciones nuevas basadas en cruzamiento y mutación.
* Población intermedia de padres que solo vive una (1) generación.

## Repaso

Antes de profundizar en los operadores avanzados, repasemos los componentes fundamentales que comparte todo Algoritmo Evolutivo: la **función de fitness**, la **población**, la **diversidad** y los **criterios de inicio y término**.

## Función de Fitness

* Función de evaluación.
* Mide la calidad de los **genotipos** (la codificación interna de un individuo).
* Mide la adaptación (_fit_, ajuste) de los individuos al ambiente (evolución del sistema).
* Se calcula decodificando los genotipos al espacio de los **fenotipos** (la solución expresada en el dominio del problema).

## Población

* Multiconjunto que representa una selección de candidatos a solución.
   * Multiconjunto: Conjunto que puede tener valores repetidos.
<br><br>
* Individuos estáticos $\rightarrow$ **no** cambian.
* Poblaciones dinámicas $\rightarrow$ **evolucionan**.
* Su tamaño, $pop$, es generalmente constante (no varía en ejecución).
* Base para la aplicación de **operadores evolutivos**.

### Diversidad

* Medida de dispersión de diferentes soluciones presentes en la población.
* Busca controlar la exploración/explotación de la evolución.
<br><br>
* Métricas:
    * Cantidad de valores de funciones de fitness
    * Cantidad de diferentes fenotipos.
    * Cantidad de diferentes genotipos.
    * Medidas estadísticas (entropía)

## Criterios de Inicio

* Para iniciar un EA, se debe contar con una **población inicial**.
* Esta población puede ser generada con procesos heurísticos.
* Se recomienda partir con una **diversidad** alta.
* Individuos $\rightarrow$ aleatorios $\rightarrow$ **población randomizada**.
<br><br>
<center><img src="images/initialpop.png" alt="Generación de la población inicial" width="90%" align="center"/></center>

## Criterios de Término

* Como gran parte de las metaheurísticas, los EAs no tienen un criterio fijo de detención.
    * Se programan para ir pasando por soluciones de calidad diversa
    * Pueden estar iterando eternamente.
<br><br>

* Criterios:
    * **Convergencia**: % de la población con mismo fitness (pérdida de diversidad)
    * Tiempo sin mejoras en la adaptación generacional.
    * Total de evaluaciones fitness hechas. 
    * Número de Iteraciones.
    * Tiempo de ejecución.

# Algoritmos basados en GA

La fama de los algoritmos genéticos y la programación genética radica en que son la base para diversos algoritmos evolutivos que usan sus principios siendo modificados en algunos de los siguientes elementos:
* Operadores de Selección.
* Operadores de Recombinación.
* Operadores de Mutación.
* Operadores de Supervivencia. 

## Tipos de Operadores de Selección

Recordemos:
> Operadores que permiten transmitir descendencia genética desde la generación actual.

### Selección Aleatoria
> Selección aleatoria uniforme.

Sea $P$ una población con $|P|$ individuos, y $N$ el conjunto de $n$ padres a seleccionar. Podemos definir $N$ como:<br>

$$ N = \{ p \in P\ |\ rand_i(P) = n \land |N_i| < n \} $$

con:
* $rand()$: Función uniforme aleatoria de selección de padres.
* $rand_i()$: i-ésimo llamado a la función $rand()$.
* $|N_i|$: Cardinalidad del conjunto $N$ al llamar a $rand_i()$.

<center><img src="images/random-selection.png" alt="" width="60%" align="center"/></center>

### Selección por Torneos
>Selección de $n$ padres por torneos basados en fitness (todos contra todos). 

#### Torneos Simples
Sea $P$ una población de $|P|$ individuos, y $T$ un torneo definido como un conjunto de $t$ participantes. Podemos definir $T$ como:<br>
$$ T = \{ t \in P\ |\ rand_i(P) = t \land |T_i| < m \} $$

Luego, los padres seleccionados serán:<br>
$$N = best(T, n, mode)$$

con:
* $mode$: modo de torneo (llaves, round robin, etc.)
* $best(T, n, mode)$: función que selecciona los $n$ mejores participantes de un torneo $T$.

**Los padres pueden ser escogidos en 1 torneo o en $n$ torneos diferentes.**

<center><img src="images/tournament-selection.png" alt="" width="60%" align="center"/></center>

#### Torneos Dobles
> Selección de $n$ padres por $n$ torneos diferentes sin repetir participantes.

Dado un conjunto $\mathcal{T}$ de $n$ torneos simples $T_i$. 

$$ N = \{best(T_i, 1, mode) \in  P\ |\ T_i \cap T_j = \varnothing \land \forall j \neq i \}$$


### Selección Proporcional a Fitness
> Criterios basados en el nivel de adaptación de los individuos previo a la selección.

#### Selección por Ranking

Los individuos son clasificados por ranking para luego escoger $n$ padres.

Los criterios de selección son:
* Elitista: $n$ mejores.
* Pesimista: $n$ peores.
* Aleatoria: $n$ al azar.
* Combinación de las anteriores.

<center><img src="images/ranking-selection.png" alt="" width="50%" align="center"/></center>

#### Selección por Ruleta

A cada individuo se le asigna una probabilidad de selección proporcional a su valor de fitness respecto al total de la población. Para un individuo $i$ con fitness $f_i$:

$$ p_i = \frac{f_i}{\sum_{j=1}^{|P|} f_j} $$

Así se obtiene una selección aleatorizada donde los mejores tienen mayor probabilidad de ser escogidos, pero sin descartar por completo a los menos aptos (se preserva algo de diversidad).

<center><img src="images/roulettewheel.png" alt="" width="35%" align="center"/></center>

#### Selección por Muestreo Estocástico Universal (SUS)

El **Muestreo Estocástico Universal** (_SUS, Stochastic Universal Sampling_) es una extensión de la selección por ruleta que utiliza $n$ punteros fijos y equidistantes —uno por cada padre a seleccionar— en lugar de girar la ruleta $n$ veces.

Con un único "giro" se eligen todos los padres a la vez, lo que **reduce la varianza** del muestreo y evita que un individuo muy apto acapare la selección. La distancia entre punteros consecutivos es constante e igual a $1/n$, con $n$ el número de padres a seleccionar.

<center><img src="images/sus.png" alt="" width="50%" align="center"/></center>

## Tipos de Operadores de Mutación
Recordemos: 
> Operadores que permiten modificar elementos puntuales de individuos de la población.

### Mutación Aleatoria

Con cierta probabilidad $p_m$, se cambia los valores de uno o algunos de los genes de un individuo.
Se aplican generalmente a la descendencia reciente generada por recombinación.

#### Mutación a Gen
Dada la selección aleatoria de un gen $g$, con probabilidad $p_m$, se cambia el valor de éste por uno de su dominio, seleccionado al azar.

<center><img src="images/random-mutation.png" alt="" width="45%" align="center"/></center>

#### Mutación Uniforme
Por cada gen $g_i$, con probabilidad $p_{m_i}$, se cambia su valor por uno de su dominio, seleccionado al azar.

<center><img src="images/random-uniform-mutation.png" alt="" width="80%" align="center"/></center>

### Mutación por Permutación

Con probabilidad $p_m$, se generan modificaciones en el orden de los genes de un individuo.

Existen variados tipos:
* Mutación por inserción.
* Mutación por intercambio.
* Mutación por mezcla/revuelta.

<center><img src="images/permutation-mutation.png" alt="" width="75%" align="center"/></center>

### Mutación Gaussiana
Con probabilidad $p_m$, se genera una perturbación gaussiana sobre un gen $g$ con valor $x_g \in [a,b]$. Es decir, se muestrea sobre una Distribución Gaussiana (Normal) $\mathcal{N}(x_g,\sigma)$ un valor que se añade al gen y luego se recorta (_clamp_) al dominio $[a,b]$:

$$ x'_g = \min(\max(\mathcal{N}(x_g, \sigma), a), b) $$

$$ x'_g =  x_g + \sigma \cdot \mathcal{N}(0,1) $$

Se aplica al igual que en la Mutación Aleatoria:
* Mutación a Gen
* Mutación Uniforme Gaussiana

<center><img src="images/gaussian-mutation-parameter.png" alt="" width="75%" align="center"/></center>

## Tipos de Operadores de Recombinación

Recordemos
> Mezcla de la información genética de dos padres en uno o dos hijos.


### Recombinación por puntos

Recombinación intercalada entre padres, basada en puntos de intercambio, para generar una descendencia de máximo $2$ individuos.

#### Recombinación a 1 punto.

A partir de un punto aleatoriamente generado, para ambos padres, se intercambian la información contenida entre ellos.

<center><img src="images/one-point-crossover.png" alt="" width="80%" align="center"/></center>

#### Recombinación a $k$ puntos.

A partir de $k$ puntos aleatoriamente generados, para ambos padres, se intercambia la información contenida entre ellos, de forma intercalada.

<center><img src="images/3-point-crossover.png" alt="Recombinación a k puntos (k=3)" width="80%" align="center"/></center>

### Recombinación Uniforme

Dada una cantidad de $k$ de padres, se genera una cantidad $l$ de hijos (generalmente 1), donde cada gen corresponde a un componente elegido aleatoriamente uniforme desde los padres.

#### Recombinación por 2 padres

<center><img src="images/uniform-crossover.png" alt="" width="80%" align="center"/></center>

### Recombinación por Permutación

Recombinaciones que alteran el orden de los genes, y son aplicables generalmente a representaciones cuyo orden no implica invalidez de los individuos (por ejemplo, permutaciones de ciudades en el TSP).

#### Recombinación por Mapeo Parcial (PMX)

El **cruzamiento por mapeo parcial** (_PMX, Partially Mapped Crossover_) genera una división por $k$ puntos de cruzamiento, donde se intercambia una o varias secciones entre los padres para generar hasta 2 hijos incompletos.

El resto de los valores se mapea según la relación de orden entre las zonas intercambiadas; los valores que no generan conflicto mantienen su lugar. Así se garantiza que cada hijo siga siendo una **permutación válida** (sin elementos repetidos).

<center><img src="images/partially-mapped-crossover.png" alt="" width="50%" align="center"/></center>

#### Recombinación por Orden (OX)
El **cruzamiento por orden** (_OX, Order Crossover_) genera una división por $k$ puntos de cruzamiento, donde se intercambia una o varias secciones entre los padres para generar hasta 2 hijos incompletos.

A diferencia de `PMX`, el resto de los valores se completa según el **orden** en que aparecen en el padre del cual no proviene la sección recibida, preservando la validez de la permutación.

<center><img src="images/order-crossover.jpg" alt="" width="70%" align="center"/></center>

## Tipos de Operadores de Supervivencia
> Operadores que emulan la adaptabilidad al ambiente de la generación actual y/o su descendencia.

### Elitismo

* Capacidad de mantener en la generación al o a los mejores individuos de la generación anterior respecto a su adaptación (fitness).
* Variantes:
    * Mantener al mejor generador de hijos.
    * Mantener al mejor fenotipo.
    * Mantener al mejor genotipo.

### Ranking de Fitness
* Selección de $n \leq size(pop)$ individuos ordenados según su valor de fitness.
* En el ranking participan padres e hijos.
* El tamaño de $n$ depende de otros criterios de supervivencia.

### Edad de Individuos
* Eliminación de un individuo según su edad.
* Simula la condición finita de permanencia de un individuo de una especie.
* Se define un parámetro de **máxima cantidad de generaciones** para los individuos.

# Derivaciones... Algoritmos Bioinspirados

Los Algoritmos Genéticos son solo una rama de un universo más amplio de **metaheurísticas bioinspiradas**: técnicas que trasladan mecanismos observados en la naturaleza a la resolución de problemas de optimización.

Mientras los GA se inspiran en la **evolución** de las especies, otras familias imitan el comportamiento **colectivo** de organismos simples (_swarm intelligence_ o inteligencia de enjambre). A continuación revisamos dos de las más influyentes:

* **ACO** (_Ant Colony Optimization_): inspirada en el forrajeo de las hormigas.
* **PSO** (_Particle Swarm Optimization_): inspirada en el vuelo coordinado de bandadas y cardúmenes.

# Ant Colony Optimization (ACO)

Algoritmo bioinspirado en la auto-organización de las hormigas _Linepithema humile_ durante el proceso de búsqueda de comida (Dorigo, 1992).

* Las hormigas son **agentes** que construyen una solución.
* Cooperan entre sí para ir detectando mejores soluciones $\rightarrow$ convergencia.
* Poseen manejo de memoria temporal $\rightarrow$ **feromona**.
* Poseen retroalimentación positiva y negativa.

## Ideas Básicas

* Feromona $\rightarrow$ memoria temporal $\rightarrow$ cómo se encontró una solución.
* Feedback positivo $\rightarrow$ reforzar buenas rutas $\rightarrow$ `feromona++`.
* Feedback negativo $\rightarrow$ disipación temporal de la memoria $\rightarrow$ `feromona--`.
* Equilibrio del feedback $\rightarrow$ explotar y explorar.
* Cooperativo $\rightarrow$ lo que hace una hormiga afecta a las otras.

<center><img src="images/aco.jpg" alt="" width="80%" align="center"/></center>

## Ant System (AS)


Dado un grafo $G = (V,A)$ con un conjunto $V$ de nodos o vértices, y un conjunto $A$ de enlaces o aristas.
* $b_i(t)$: el número de hormigas en el nodo $i$, con $i = \{1,2,\dots, |V|\}$.
* $m = \sum_{i}^{|V|} b_i(t) \rightarrow$ el número total de hormigas (agentes).

### Feromona: Actualización
La intensidad de la feromona $\tau_{ij}(t+z)$ en la arista $(i,j)$ en el instante $(t+z)$ está dada por:<br>

$$ \tau_{ij}(t+z) = (1-\rho)\, \tau_{ij}(t) + \Delta \tau_{ij}(t+z)$$

donde $\rho$ es el coeficiente de evaporación de la feromona y $\Delta \tau_{ij}(t+z)$ es la cantidad de feromona depositada en el arco $(i,j)$ en el instante $(t+z)$.

_Nota: no importa si el grafo es simétrico o si está completamente conectado._

### Feromona: Incremento
La cantidad de feromona dejada en un arco en un determinado instante está dado por:


$$ \Delta \tau_{ij}(t+z)= \left\{ \begin{array}{ll} \frac{Q}{L_k} & \text{si la $k$-ésima hormiga usa el arco } (i,j)    \\ 0 & \text{e.o.c} \\ \end{array} \right. $$

donde $Q$ es una constante (parámetro) y $L_k$ es el valor de la solución encontrada por la $k$-ésima hormiga.

* Nota: La intensidad de feromona inicial es casi cero:  $\tau_{ij}(0) \rightarrow 0$.

### Regla de Transición: ¿Cómo avanzan las hormigas?

La probabilidad de que un agente, avance de un nodo $i$ al nodo $j$ está dada por: 

$$ P_{ij}^{k}(t)= \left\{ \begin{array}{ll} \frac{[\tau_{ij}(t)]^\alpha [\eta_{ij}]^\beta}{\sum_{h\ \in \text{ permitido}_k(t)} [\tau_{ih}(t)]^\alpha [\eta_{ih}]^\beta} & \text{si nodo $j \in $ permitido$_k(t)$}    \\ 0 & \text{e.o.c} \\ \end{array} \right. $$

Donde:
* $\eta_{ij}$ es una heurística seleccionada respecto al problema a resolver; y 
* $\alpha$ y $\beta$ son parámetros que permiten decidir como se reparte el peso de la probabilidad entre la feromona y la heurística seleccionada.

### AS Elitista

Variación de la función de actualización de la feromona, considerando a la hormiga que encuentra la solución como un super depositador de feromona, vale decir, amplificar su efecto. 

$$ \tau_{ij} = (1-\rho) \tau_{ij}(t) + \Delta \tau_{ij}(t+z) + e \Delta \tau_{ij}^{e}(t)$$

donde:

$$ \Delta \tau_{ij}^{e}(t)= \left\{ \begin{array}{ll} \frac{Q}{L^+} & \text{si la $(i,j) \in T^+$} \\ 0 & \text{e.o.c} \\ \end{array} \right. $$

con $L^+$ es el valor de la mejor solución, y $T^+$ es la configuración de la mejor solución.

## Ant Colony System (ACS)

El **Ant Colony System** (_ACS_, Dorigo y Gambardella, 1997) es una mejora del Ant System que acelera la convergencia y explota mejor la información aprendida. Introduce tres cambios clave:

* **Regla de transición pseudo-aleatoria proporcional:** con probabilidad $q_0$ la hormiga elige de forma *greedy* el mejor arco disponible (explotación); con probabilidad $1-q_0$ usa la regla probabilística del AS (exploración).
* **Actualización global de feromona:** solo la **mejor** hormiga (la del mejor recorrido) deposita feromona, concentrando el refuerzo en las soluciones prometedoras.
* **Actualización local de feromona:** cada vez que una hormiga cruza un arco evapora un poco de su feromona, favoreciendo que otras hormigas exploren rutas alternativas dentro de la misma iteración.

# Particle Swarm Optimization (PSO)

El **Particle Swarm Optimization** (_PSO_, Kennedy y Eberhart, 1995) es una metaheurística de **inteligencia de enjambre** inspirada en el movimiento coordinado de bandadas de aves y cardúmenes de peces. A diferencia de los GA, no usa cruza ni mutación: las soluciones (llamadas **partículas**) **se mueven** por el espacio de búsqueda guiadas por su propia experiencia y la del enjambre.

## Idea básica

Cada partícula $i$ tiene una **posición** $x_i$ (una solución candidata) y una **velocidad** $v_i$. En cada iteración recuerda su mejor posición personal $p_i$ (_pbest_) y conoce la mejor posición global del enjambre $g$ (_gbest_). La velocidad y la posición se actualizan según:

$$ v_i(t+1) = w\, v_i(t) + c_1 r_1\, (p_i - x_i(t)) + c_2 r_2\, (g - x_i(t)) $$

$$ x_i(t+1) = x_i(t) + v_i(t+1) $$

donde:
* $w$ es la **inercia** (controla la exploración),
* $c_1$ y $c_2$ son los coeficientes **cognitivo** y **social**,
* $r_1, r_2 \sim U(0,1)$ son factores aleatorios.

## ACO y PSO como metaheurísticas bioinspiradas

Tanto **ACO** como **PSO** pertenecen a la familia de la **inteligencia de enjambre** (_swarm intelligence_): muchos agentes simples que, mediante interacción local y cooperación, producen un comportamiento global inteligente. Junto con los Algoritmos Evolutivos comparten el equilibrio entre **exploración** y **explotación** y la ausencia de garantías de optimalidad, pero difieren en su metáfora:

| Enfoque | Inspiración | Memoria / Comunicación | Representación típica |
| :-- | :-- | :-- | :-- |
| **GA / EA** | Evolución y selección natural | Población de individuos | Binaria, real, permutación |
| **ACO** | Forrajeo de hormigas | Feromona en el grafo (indirecta, _estigmergia_) | Camino en un grafo |
| **PSO** | Bandadas / cardúmenes | pbest y gbest (directa) | Vector real |

| :-- | :-- |
| **Selección** | Torneo (simple/doble), ranking, ruleta y SUS deciden qué padres se reproducen según su fitness. |
| **Recombinación** | 1 punto, $k$ puntos y uniforme para representaciones estándar; PMX y OX para permutaciones (preservan validez). |
| **Mutación** | Aleatoria (a gen / uniforme), por permutación (inserción, intercambio, mezcla) y gaussiana para variables reales. |
| **Supervivencia** | Elitismo, ranking de fitness y edad de los individuos determinan quién pasa a la siguiente generación. |
| **ACO** | Metaheurística de hormigas basada en feromona; Ant System, AS elitista y ACS aplicados al TSP. |
| **PSO** | Inteligencia de enjambre con partículas guiadas por pbest y gbest. |
| **Bioinspiradas** | GA, ACO y PSO equilibran exploración y explotación imitando procesos naturales. |
