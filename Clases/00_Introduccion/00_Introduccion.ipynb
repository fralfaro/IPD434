## 🎯 Objetivos de Aprendizaje

Al finalizar esta clase, el estudiante será capaz de:

1. **Conocer** la estructura, la metodología y la forma de evaluación del curso IPD434.
2. **Comprender** qué es el *Soft Computing* y en qué se diferencia del *hard computing*.
3. **Identificar** las áreas que cubre el curso: Fuzzy Logic, Computación Evolutiva, Redes Neuronales y especialidades (RL/DRL, GAN, Transformers, ACO, PSO).
4. **Ubicar** cada módulo dentro de un mapa mental del campo y entender por qué es relevante.
5. **Reconocer** las herramientas de trabajo del curso: Python, Jupyter y las librerías principales.

> 💡 **Prerrequisitos:** Ninguno — este es el punto de partida del curso.

# Estructura del Curso

## Evaluación del curso

$$ \text{NF} = 0.5\cdot\text{C} + 0.3\cdot\text{PR} + 0.2\cdot\text{LP}$$

Donde
- $\text{NF}$: Nota Final
- $\text{C}$: Certamen
- $\text{PR}$: Proyecto
- $\text{LP}$: Lectura y Presentación de Paper

### Certamen

Se evaluán los tópicos bases del curso:
* Fuzzy Logic.
* Redes Neuronales Artificiales.
* Computación Evolutiva.

### Proyecto

* Resolución de Problema de Alta Complejidad con técnicas de Soft Computing.
    * Telemáticos: Problemas en Telecomunicaciones.
    * Informáticos: Problemas en Ciencias de la Computación.

* Short-Paper: Estado del Arte, Técnica, Resultados
* Presentación Final.

* CCCV: Centrado en Redes Neuronales.
* CSSJ: Centrado en Computación Evolutiva y Algoritmos Bío-Inspirados.

### Lectura y Presentación de Paper

Selección de un Problema Abordado con Técnicas de Soft Computing.

* Fuzzy Logic.
* Redes Neuronales.
* Computación Evolutiva.
* Algoritmos Bío-Inspirados.
