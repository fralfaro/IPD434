## 🚀 Propósito y Motivación del Curso

Muchos de los problemas más interesantes de la ingeniería no tienen una solución exacta y cerrada: reconocer un rostro, conducir un vehículo, traducir un idioma o encontrar rutas óptimas en una red son tareas donde la información es **incompleta, imprecisa o ruidosa**, y donde buscar la solución *perfecta* resulta demasiado costoso. El **Soft Computing** nace precisamente para abordar esta clase de problemas: en lugar de exigir precisión absoluta, tolera la imprecisión y la incertidumbre para obtener soluciones **robustas, flexibles y de bajo costo computacional**.

Este curso te entregará una caja de herramientas para atacar problemas complejos donde los métodos clásicos fallan o son inviables:

- 🔮 **Razonar con imprecisión** usando lógica difusa (*fuzzy logic*).
- 🧬 **Optimizar** espacios de búsqueda enormes con algoritmos evolutivos y bioinspirados.
- 🧠 **Aprender de los datos** con redes neuronales y aprendizaje profundo (*deep learning*).

> 🎯 **La meta:** que al terminar el curso seas capaz de *elegir* la técnica de Soft Computing adecuada para un problema real, *implementarla* en Python y *evaluar* críticamente sus resultados.

## 🧠 ¿Qué es Soft Computing?

El término **Soft Computing** (computación *flexible* o *suave*) fue acuñado y popularizado por **Lotfi Zadeh** en la década de 1990. Se define como una familia de técnicas que **explotan la tolerancia a la imprecisión, la incertidumbre y la verdad parcial** para lograr soluciones tratables, robustas y de bajo costo, imitando la notable capacidad de la mente humana para tomar buenas decisiones con información incompleta.

Se contrapone al **Hard Computing** (computación *dura* o convencional), que exige un modelo matemático preciso y datos exactos, y produce resultados exactos, pero que suele volverse frágil o intratable frente a problemas reales mal definidos.

### Soft Computing vs. Hard Computing

| Característica          | Hard Computing                 | Soft Computing                              |
|------------------------|--------------------------------|---------------------------------------------|
| **Modelo**             | Preciso, analítico             | Aproximado, basado en datos y heurísticas   |
| **Datos**              | Exactos y completos            | Imprecisos, ruidosos, incompletos           |
| **Lógica**             | Binaria (verdadero/falso)      | Multivaluada / difusa (grados de verdad)    |
| **Resultado**          | Exacto y determinista          | Aproximado, *suficientemente bueno*         |
| **Tolerancia al error**| Baja (frágil)                  | Alta (robusto)                              |
| **Ejemplos**           | Ordenamiento, álgebra lineal   | Lógica difusa, algoritmos genéticos, redes neuronales |

### Los pilares del Soft Computing

Este curso se organiza en torno a los tres pilares clásicos del campo, junto con sus extensiones modernas:

- **Lógica Difusa (*Fuzzy Logic*)** — modela el *razonamiento aproximado* y la vaguedad del lenguaje humano (por ejemplo, "el agua está *tibia*").
- **Computación Evolutiva y Bioinspirada** — *optimiza* imitando la evolución natural (algoritmos genéticos) y el comportamiento colectivo (colonias de hormigas, enjambres de partículas).
- **Redes Neuronales** — *aprenden* patrones directamente de los datos y son la base del aprendizaje profundo actual.

> 💡 Estos paradigmas no son excluyentes: con frecuencia se **combinan** en sistemas *híbridos* (por ejemplo, neuro-difusos o neuroevolutivos) para aprovechar las fortalezas de cada uno.

## 📜 Panorama Histórico

El Soft Computing es heredero de varias líneas de investigación que, durante el siglo XX, buscaron construir máquinas capaces de razonar, aprender y adaptarse. Conocer a sus pioneros ayuda a entender de dónde proviene cada módulo del curso.

| Año  | Hito                                                        | Protagonista(s)               | Módulo relacionado          |
|------|-------------------------------------------------------------|-------------------------------|-----------------------------|
| 1943 | Modelo de neurona artificial                                | McCulloch y Pitts             | Redes Neuronales            |
| 1950 | *Computing Machinery and Intelligence* y el Test de Turing  | **Alan Turing**               | Fundamentos de la IA        |
| 1957 | **Perceptrón**, primera red entrenable                      | **Frank Rosenblatt**          | Redes Neuronales            |
| 1965 | Teoría de **conjuntos difusos**                             | **Lotfi Zadeh**               | Sistemas Difusos            |
| 1975 | **Algoritmos Genéticos**                                    | **John Holland**              | Computación Evolutiva       |
| 1986 | Popularización de *backpropagation*                         | Rumelhart, Hinton y Williams  | Redes Neuronales            |
| 1992 | *Ant Colony Optimization* (ACO)                             | Marco Dorigo                  | Algoritmos Bioinspirados    |
| 1995 | *Particle Swarm Optimization* (PSO)                         | Kennedy y Eberhart            | Algoritmos Bioinspirados    |
| 2012 | El *deep learning* despega (AlexNet)                        | Krizhevsky, Sutskever, Hinton | Redes Neuronales / CNN      |
| 2017 | Arquitectura **Transformer** (*Attention Is All You Need*)  | Vaswani et al.                | Transformers                |

### Ideas clave de la línea de tiempo

- **Alan Turing (1950)** planteó la pregunta fundamental "¿pueden pensar las máquinas?" y propuso el *Test de Turing*, sentando las bases filosóficas de la Inteligencia Artificial.
- **Frank Rosenblatt (1957)** construyó el **perceptrón**, el primer modelo neuronal capaz de *aprender* a partir de ejemplos.
- **Lotfi Zadeh (1965)** introdujo los **conjuntos difusos**, formalizando cómo razonar con conceptos vagos y grados de pertenencia.
- **John Holland (1975)** formalizó los **algoritmos genéticos**, trasladando la metáfora de la evolución darwiniana al problema de la optimización.

## 🗺️ Mapa del Curso

El curso avanza desde los **fundamentos** comunes a todo el Soft Computing hacia **especialidades** avanzadas. Primero se cubren los tres pilares (Lógica Difusa, Computación Evolutiva y Redes Neuronales), que constituyen la base evaluada en el **Certamen**; luego, según la línea (**CCCV** o **CSSJ**), se profundiza en temas de vanguardia.

### Módulos y su relación

- **Fundamentos comunes**
  - *Sistemas Difusos* — razonamiento bajo incertidumbre.
  - *Computación Evolutiva* — optimización bioinspirada.
  - *Redes Neuronales* — aprendizaje a partir de datos.
- **Especialidad CCCV — centrada en Redes Neuronales**
  - *Reinforcement Learning / Deep RL* — agentes que aprenden por recompensa.
  - *Transformers* — mecanismos de atención y modelos de secuencia.
  - *GAN y Autoencoders* — modelos generativos.
- **Especialidad CSSJ — centrada en Computación Evolutiva y Bioinspirada**
  - *Algoritmos Genéticos y Programación Genética* (GA/GP).
  - *Ant Colony Optimization* (ACO) — inteligencia de colonias.
  - *Particle Swarm Optimization* (PSO) — inteligencia de enjambres.

> 🔗 **¿Cómo se conectan?** Los tres pilares comparten una misma filosofía —*aproximar soluciones a problemas complejos*— y a menudo se combinan: la computación evolutiva puede **optimizar** los pesos o la arquitectura de una red neuronal (*neuroevolución*), y la lógica difusa puede **interpretar** las salidas de un modelo de aprendizaje.
