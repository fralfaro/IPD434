# Resolviendo problemas: Método de la Caja Negra

Es un método que sirve para clasificar los diversos tipos de problemas que necesitan ser resueltos en la vida real. Se compone de tres elementos:
* Una entrada, que representa la información de entrada a del sistema;
* Un modelo, que toma al entrada y la utiliza, transforma y/o modifica; y
* Una salida, que representa el resultado generado por el modelo.

# Optimización Combinatoria

> Campo de la matemática que busca encontrar el objeto óptimo que resuelve a un problema dentro de un conjunto finito de objetos.  

El desafío está en que, aunque el conjunto es finito, puede ser **enorme**, lo que hace imposible revisar todas las opciones una por una.

## Problemas de Optimización Combinatoria

Los problemas de Optimización Combinatoria se pueden clasificar dentro de los primeros dos problemas de tipo caja negra:

* Problemas de Optimización:
    * Modelo: Modelación Matemática (lineal, cuadrática, lógica primer orden).
    * Salida: Función de Evaluación (objetivo, fitness, etc)
    + **Entrada**: Variables, y sus valores, que optimizan la salida.

* Problemas de Modelado:
    * Entrada: Data Set. 
    * Salida: Etiquetas (conocidas y aprendidas).
    * **Modelo**: Sistema que permite etiquetar entradas en salidas con cierto grado de precisión.
    * _Puede ser re-escrito como un Problema de Optimización_:
        * Modelo: Modelado estadístico (i.e, regresión)
        * Salida: Error Cuadrático Medio, Precisión, Coeficiente de Determinación.
        * **Entrada**: Variables y valores que optimizan la salida.   

## Problemas de Búsqueda

* Resolver un problema muchas veces significa **explorar un espacio de posibilidades**.
* Ese espacio, llamado **espacio de estados o espacio de búsqueda**, contiene todas las soluciones posibles.
* Aunque es finito, suele ser tan grande que no es factible revisar cada opción una por una.
* Por eso, los problemas de **optimización combinatoria** pueden entenderse como **problemas de búsqueda**: buscamos la mejor solución dentro de un conjunto inmenso de candidatos.

## Proceso de Búsqueda

**Definición:**

> Encontrar, evaluar y aceptar soluciones como candidatas para resolver un problema, dentro de todas las soluciones posibles.

1. **Encontrar:**

   * Identificar el **espacio de búsqueda**, es decir, todas las combinaciones posibles de las variables de decisión.

2. **Evaluar:**

   * Usar una **función de evaluación** que mida la calidad de cada solución (también llamada función objetivo o de fitness).

3. **Aceptar:**

   * Determinar si la solución cumple con las **restricciones** definidas del problema.

## Tipos de Problemas de Búsqueda

Los problemas de búsqueda pueden clasificarse según el objetivo: encontrar la mejor solución, cumplir restricciones, o ambas a la vez.

<center><img src="images/cop.png" alt="" width="50%" align="center"/></center>

### Problema de Optimización Libre (_Free Optimisation Problem, FOP_)
> Se busca una solución que optimiza un función de evaluación sin restricciones (más que la estructura de la misma).

### Problema de Satisfacción de Restricciones (_Constraint Satisfaction Problem, CSP_)
> Se acepta una solución que cumple con un grupo de restricciones que no optimiza alguna función de evaluación.

### Problema de Optimización y Satisfacción de Restricciones (_Constraint Satisfaction and Optimisation Problem, CSOP_)
##### a.k.a. Problema de Optimización Restringido (_Constrained Optimisation Problem, COP_)
> Se busca una solución que optimiza una función de evaluación y cumple con un grupo de restricciones adosadas a ella. 

### Ejemplo: Problema de las N-Reinas

Problema de Satisfacción de Restricciones (CSP), de tipo puzzle matemático, de complejidad **NP-Complete**, que establece lo siguiente: 

> Ubique $n$ reinas en un tablero de ajedrez de dimensiones $n \times n$ de tal forma que no habrá dos reinas que hagan jaque entre ellas. 

<center><img src="images/nqueens.png" alt="" width="30%" align="center"/></center>

### N-Reinas como FOP

En este caso no hay restricciones explícitas, simplemente buscamos la mejor solución posible según una función de evaluación.

* Representación: Tablero de ajedrez de tamaño $n \times n$. 
</br></br>
* Espacio de Búsqueda:
    * $S$: Conjunto de todas las configuraciones del tablero con $n$ reinas en ella. 
</br></br>
* Función de Evaluación: 
    * $f(s)$: Cantidad de reinas libres de check en el tablero 
</br></br>
* Soluciones:
    * $ S^* = \{ s \in S ~ | ~ f(s) = n \}$
    * $ S^* \subseteq S$

### N-Reinas como CSP

Aquí solo se desea cumplir con una restricción: ninguna reina puede estar en jaque.

* Representación: Tablero de ajedrez de tamaño $n \times n$.
</br></br>
* Espacio de Búsqueda:
    * $S$: Conjunto de todas las configuraciones del tablero con $n$ reinas en ella. 
</br></br>    
* Restricción (como predicado): 
    * ```is_checked(s)```$ = \left\{ \begin{array}{ll} \texttt{true} & \text{si } s \in S \text{ tiene dos reinas en check} \\ 
                              \texttt{false} & \text{en caso contrario} \end{array} \right.$
</br></br>    
* Soluciones:
    * $ S^* = \{ s \in S ~ | $ ```is_checked(s)```$ = \texttt{false} \}$
    * $ S^* \subseteq S$

### N-Reinas como CSOP

También es posible combinar ambos enfoques: cumplir restricciones y al mismo tiempo optimizar la solución.

* Representación: Tablero de ajedrez de tamaño $n \times n$.
</br></br>    
* Espacio de Búsqueda:
    * $S$: Conjunto de todas las configuraciones del tablero con $n$ reinas en ella. 
</br></br>    
* Restricción (como predicado): 
    * ```column(s)```$ = \left\{ \begin{array}{ll} \texttt{true} & \text{si } s \in S \text{ tiene dos reinas en una misma columna} \\ 
                              \texttt{false} & \text{en caso contrario} \end{array} \right.$
</br></br>
    * ```row(s)```$ = \left\{ \begin{array}{ll} \texttt{true} & \text{si } s \in S \text{ tiene dos reinas en una misma fila} \\ 
                              \texttt{false} & \text{en caso contrario} \end{array} \right.$
</br></br>
* Función de Evaluación: 
    * $D(s)$: Cantidad de reinas en check por diagonales. 
</br></br>
* Soluciones:
    * $ S^* = \{ s \in S ~ | $ ```column(s)```$ = \texttt{false}~ \land $ ```row(s)```$ = \texttt{false}~ \land min(D(s))\}$
    * $ S^* \subseteq S$

* La representación de la solución, o modelo, juega un rol fundamental en como se aborda y soluciona el problema.
* Puede mejorar drásticamente la capacidad de resolución del problema.
* **Out-of-the-box thinking**.

Además de definir el problema como FOP, CSP o CSOP, la **representación de la solución** es clave.

* En nuestro ejemplo, usar directamente el tablero es poco eficiente.
* Una alternativa es asumir **una reina por fila y columna**, reduciendo el espacio de búsqueda.
* Esto permite representaciones más compactas y fáciles de evaluar.

### Tiene que haber obligatoriamente una reina en cada fila y en cada columna. Version 1.

* Representación: Vector de enteros de tamaño $n$, dónde cada elemento representa secuencialmente a cada columna de un tablero de ajedrez de tamaño $n \times n$. El valor en cada elemento corresponde a la fila en la que se ubica una reina.
</br></br>    
* Espacio de Búsqueda:
    * $S$: Conjunto de todos los vectores enteros válidos representando un tablero de ajedrez de $n \times n$.  
* Restricción (como predicado): 
    * ```all_different(s)```$ = \left\{ \begin{array}{ll} \texttt{true} & \text{si } s \in S \text{ no tiene dos elementos del vector que tengan el mismo valor} \\ 
                              \texttt{false} & \text{en caso contrario} \end{array} \right.$  
* Función de Evaluación: 
    * $D(s)$: Cantidad de reinas en check por diagonales.     
* Soluciones:
    * $ S^* = \{ s \in S ~ | $ ```all_different(s)```$ = \texttt{false}~ \land min(D(s))\}$
    * $ S^* \subseteq S$

### Tiene que haber obligatoriamente una reina en cada fila y en cada columna. Version 2.

* Representación: Vector de enteros de tamaño $n$, dónde cada elemento representa secuencialmente a cada columna de un tablero de ajedrez de tamaño $n \times n$. El valor en cada elemento corresponde a la fila en la que se ubica una reina.
* Espacio de Búsqueda:
    * $S$: Conjunto de todas los vectores enteros que no repiten valores.  
* Función de Evaluación: 
    * $D(s)$: Cantidad de reinas en check por diagonales.      
* Soluciones:
    * $ S^* = \{ s \in S ~ | min(D(s))\}$
    * $ S^* \subseteq S$

# Paradigmas de Búsqueda
### a.k.a. Hard Computing vs Soft Computing en Optimización Combinatoria

Existen distintos paradigmas para afrontar encontrar soluciones en el espacio de búsqueda, las podemos resumir en las siguientes:

### Aproximación Sistémica vs Búsqueda Local

* Aproximación Sistémica:
    * Completitud.
    * Búsqueda sistémica sobre el espacio de búsqueda, garantizando: 
        * Solución óptima, o que
        * No existe solución.
    * Computacionalmente costoso. 

* Búsqueda Local:
    * Incompletitud.
    * Búsqueda parcial sobre diferentes sub-espacios del espacio de búsqueda. No puede garantizar:
        * Solución óptima.
        * Solución factible.
        * No existe solución.
    * Computacionalmente rápida.



### Búsqueda Constructiva vs Búsqueda Perturbativa.
* Búsqueda Constructiva
    * Encontrar la solución desde un inicio vacío.
    * Generalmente, soluciones incompletas. 

* Búsqueda Perturbativa
    * Encontrar la solución modificando una existente.
    * Generalmente, mejoras sobre un candidato de baja calidad. 

## Heurísticas

* Son **reglas prácticas o criterios** para guiar la búsqueda de soluciones.
* Se apoyan en el **conocimiento del problema** para tomar decisiones rápidas.
* Generalmente son **miopes**: evalúan solo la ganancia inmediata sin considerar el impacto global.
* **Ejemplo:**
  * **Algoritmo Greedy** : construye una solución paso a paso eligiendo siempre la opción localmente mejor según la heurística.

# Metaheurísticas

> **Proceso automático para buscar una buena solución, ya sea construyéndola o reparándola, respecto a un conjunto de candidatos finito pero intratable por su tamaño.**

* Proporcionan **estrategias globales** para acercarse rápidamente a un óptimo (aunque no siempre garantizan el mejor).
* No son reglas específicas del problema (heurísticas), sino un **marco genérico** que puede aplicarse a distintos contextos.
* Suelen combinar **búsqueda local** (mejorar soluciones en un área reducida) con **búsqueda perturbativa** (explorar nuevas regiones del espacio).

## Conceptos Claves

### Búsqueda Local
> Encontrar una solución óptima en una porción reducida del Espacio de Búsqueda.

### Búsqueda Perturbativa.
> Modificar un candidato a solución conocido para obtener uno nuevo. 

### Explotación o Intensificación
> Concentrar la búsqueda en la porción de espacio de búsqueda visitado.

### Exploración o Diversificación
> Ampliar la búsqueda a un fragemento más grande del espacio de búsqueda. 

### Explotación vs Exploración
> Ambas son relevantes, debe haber un equilibrio que varía entre problemas.
* Si nos concentramos en Explotar:
    * $\rightarrow$ estancamiento en óptimo local de forma prematura.
* Si nos concentramos en Explorar:
    * $\rightarrow$ búsqueda no convergerá a solución alguna. 

## ¿Como búscan las Meta-heurísticas clásicas?

### Candidato
> Solución construida o obtenida de otro proceso que será evaluada y modificada en pos de mejoras.

### Movimiento: Perturbación
> Cambio que se le puede aplicar al candidato para generar nuevas soluciones.

### Vecindario
> Conjunto de nuevas soluciones generadas desde el candidato a aplicar el movimiento en las distntas opciones posibles.

### Criterio de Selección
> Directriz para seleccionar al siguiente candidato en la búsqueda.
* Alguna mejora: Primer vecino que mejore la función de evaluación.
* Mejor mejora: Vecino que maximice la mejora de la función de evaluación.

### Criterios de Reinicio
> Criterios para evitar estancamiento en óptimo local.
* Reiniciar luego de ciertas cantidad de iteraciones.
* Reiniciar desde un nuevo candidato inicial. 

### Criterio de Término
> Criterio de término de la meta-heurística
* No encontrar un mejor vecino. 
* Cumplir número de interaciones.
* Tiempo máximo. 

### Meta-heurísticas clásicas

* **Hill-Climbing:** Parte de una solución inicial y la mejora iterativamente mientras encuentra vecinos mejores. Puede quedar atrapado en óptimos locales.
* **Iterative Local Search:** Repite múltiples búsquedas locales desde diferentes puntos para escapar de soluciones pobres.
* **Tabu Search:** Búsqueda local que evita volver a soluciones recientes, gracias a una memoria de "movimientos prohibidos" (tabú).
* **Simulated Annealing:** Inspira en el enfriamiento del metal; acepta soluciones peores con cierta probabilidad para escapar de óptimos locales, reduciendo ese margen con el tiempo.
* **GRASP (Greedy Randomized Adaptive Search Procedure):** Construye soluciones aleatorias guiadas por criterios greedy y luego las mejora con búsqueda local.

## ¿Y el multi-objetivo?

En muchos problemas no buscamos optimizar **un solo criterio**, sino **varios al mismo tiempo**, que incluso pueden entrar en conflicto (por ejemplo, minimizar costos y maximizar calidad).

* **Evaluación con pesos:** Combinar los objetivos en una sola función ponderada.
* **Bi-objetivo o Bi-criterio:** Tratar los objetivos de forma separada y buscar un compromiso entre ellos.
* **Frente de Pareto:** Conjunto de soluciones "óptimas" en el sentido de que **no se puede mejorar un objetivo sin empeorar otro**.

La figura muestra este frente:

* Los puntos azules (**bruteforce**) representan todas las soluciones evaluadas.
* Los puntos rojos corresponden a las soluciones no dominadas, que forman el **frente de Pareto**.
* El punto verde es una de las soluciones encontradas en ese frente.

## Configuración de parámetros en meta-heurísticas

Las meta-heurísticas requieren parámetros (ej. número de iteraciones, tamaño del vecindario, probabilidad de aceptar peores soluciones). La forma de configurarlos influye directamente en los resultados.

Existen dos enfoques principales:

* **Parameter Tuning (ajuste previo):**
  * Los valores se definen **antes de ejecutar** el algoritmo.
  * Generalmente a través de pruebas experimentales o recomendaciones.

* **Parameter Control (ajuste dinámico):**
  * Los valores se **modifican durante la ejecución** del algoritmo.
  * Puede hacerse de distintas formas:

    * **Deterministic:** siguen una regla fija (ej. disminuir un parámetro en cada iteración).
    * **Adaptive:** cambian en función del desempeño del algoritmo (se ajustan según resultados observados).
    * **Self-Adaptive:** los propios parámetros evolucionan junto con las soluciones.

## Representación en Algoritmos Genéticos

En los algoritmos genéticos, cada **solución real** debe traducirse a una forma que el algoritmo pueda manipular. Para eso distinguimos dos niveles:

* **Fenotipo (mundo real):**
  * La solución tal como existe en el problema original.
  * Ejemplo: una configuración del tablero en N-Reinas, un plan de horarios, un diseño de ruta.
  * Conjunto: **espacio de fenotipos**.

* **Genotipo (mundo evolutivo):**
  * La forma codificada de la solución, utilizada por el algoritmo.
  * Ejemplo: un cromosoma o cadena de genes que codifica la posición de cada reina.
  * Conjunto: **espacio de genotipos**.

* **Representación (mapeo fenotipo–genotipo):**
  * Proceso que conecta ambos mundos:
    * **Codificación:** del mundo real al evolutivo.
    * **Decodificación:** del mundo evolutivo al real.

* **Gen:**

* **Alelo:**
  * Es el **valor concreto** que toma un gen en un individuo.
  * Representa una **instanciación específica** de esa variable.

## Función de Fitness

* Es la **función de evaluación** utilizada en los algoritmos genéticos.
* Permite medir la **calidad** de cada genotipo y, por lo tanto, su capacidad de "sobrevivir" en el proceso evolutivo.
* Refleja qué tan **bien adaptado** está un individuo al problema que se quiere resolver.
* Para calcularla, normalmente se **decodifica el genotipo al fenotipo** y se mide su desempeño en el mundo real.

## Población

* Conjunto de **candidatos a solución** que el algoritmo manipula en cada generación.
* Se habla de **multiconjunto** porque pueden existir individuos repetidos.
* **Individuos estáticos:** no cambian a lo largo de la ejecución.
* **Poblaciones dinámicas:** evolucionan aplicando operadores como selección, mutación o recombinación.
* El tamaño de la población $pop$ suele mantenerse **constante** durante la ejecución.
* Es la **base** sobre la cual actúan los operadores evolutivos que impulsan la búsqueda de mejores soluciones.

### Diversidad

* Indica **qué tan distintas** son las soluciones dentro de la población.
* Es clave para mantener un equilibrio entre **exploración** (probar nuevas regiones del espacio) y **explotación** (mejorar soluciones ya conocidas).

**Formas de medirla:**

* Variación en los valores de la **función de fitness**.
* Cantidad de **fenotipos diferentes** (soluciones distintas en el mundo real).
* Cantidad de **genotipos diferentes** (cromosomas distintos).
* Métricas estadísticas como la **entropía**.

## Operadores

Son los **mecanismos principales** que permiten que una población evolucione de una generación a otra. Se dividen en tres categorías:

* **Operadores de selección de evolución:**

  * Determinan **qué individuos se convierten en padres** y transmiten su información genética a la siguiente generación.
  * Buscan favorecer a los mejores, pero manteniendo cierta diversidad.

* **Operadores de evolución:**

  * **Mutación:** introduce pequeñas variaciones aleatorias en un individuo, explorando nuevas zonas del espacio de búsqueda.
  * **Recombinación (cruzamiento):** combina información de dos padres para generar uno o más hijos, mezclando características de ambos.

* **Operadores de selección de supervivencia:**

  * Deciden **qué individuos permanecen en la población** para la siguiente generación.
  * Estrategias comunes:

    * **Elitismo:** conservar a los mejores individuos sin cambios.
    * **Fitness Ranking:** ordenar por calidad y seleccionar los mejores.
    * **Edad de individuos:** descartar a los más antiguos para mantener renovación.

## Operadores de selección de evolución.
> Operadores que permite transmitir descendencia genetica de la generación actual. 
### Selección de Padres

* Filtrar miembros de la población según su calidad, permitiendoles ser padres de la próxima generación.
* Adaptabilidad al entorno. 
* Mantener sus buenas caracteristicas.
* Busca mejoras en la función fitness de forma global.
* No se puede dejar sin chances a un individuo de baja calidad. 
<br><br>
* Probabilistica.
    * Aleatoria
    * Ranking.
    * Torneos.
    * Ruletas.

## Operadores de evolución
> Operadores que permiten modificar la población.


### Mutación

* **Perturbación** aplicada a un **padre** para generar un **hijo**.
    * Operación unaria: Sobre un solo individuo.
<br><br>
* La perturbación (movimiento) es aplicada de forma **aleatoria**.
    * Las mutaciones no son deterministas, son **estocásticas**. 
    * Las mutaciones en sistemas biológicos no pueden ser predecidas. Ej. COVID19.
        * Pueden no ocurrir.
<br><br>    
* Permite conectar distintos sectores del espacio de búsqueda.

## Operadores de selección de supervivencia
> Operadores que emulan la adaptabilidad al ambiente de la generación actual y/o su descendencia. 

### Elitismo

* Capacidad de mantener en la generación al o a los mejores individuos de la generación anterior respecto a su adaptación (fitness).
* Variantes:
    * Mantener al mejor generador de hijos.
    * Mantener al mejor fenotipo.
    * Mantener al mejor genotipo.

### Ranking de Fitness
* Selección de $n \leq size(pop)$ individuos ordenados según su valor de fitness.
* En el ranking participan padres e hijos.
* El tamaño de $n$ depende de otros criterios de supervivencia.

### Edad de Individuos
* Eliminación de individuo según su edad. 
* Simula la condicion finita de permanencia de un individuo de una especie.
* Se define un parámetro de **máxima cantidad de generaciones** para los individuos.

* Para iniciar un EA, se debe contar un una **población inicial**.
* Esta población puede ser generada con procesos heurístico.
* Se recomendienda partir con una **diversidad** alta.
* Invididuos $\rightarrow$ aleatorios $\rightarrow$ **población randomizada**.

## Criterios de Término

* Como gran parte de las metaheurísticas, los EAs no tienen un criterio fijo de detención.
    * Se programan para ir pasando por soluciones de calidad diversa
    * Pueden estar iterando eternamente.
<br><br>

* Criterios:
    * **Convergencia**: % de la población con mismo fitness (pérdida de diversidad)
    * Tiempo sin mejoras en la adaptación generacional.
    * Total de evaluaciones fitness hechas. 
    * Número de Iteraciones.
    * Tiempo de ejecución.
    
    

# Tipos de Algoritmos Evolutivos

* Si bien existen variados tipos de algoritmos evolutivos o derivados. Se reconecen las siguente 4 áreas principales:

    1. Programación Evolutiva
    1. Estrategias de Evolución
    1. Algortimos Genéticos
    1. Programación Genética
<br>
<center><img src="images/ea.png" alt="" width="90%" align="center"/></center>

### Programación Evolutiva (_EP, Evolutionary Programming_)
* Fogel et al. (1966): Simulación evolutiva $\rightarrow$ aprendizaje. 
* Adaptación: Capacidad de predecir el ambiente.
* Cada individuo es una especie diferente $\rightarrow$ no hay recombinación.

| Caracteristica | Aplicación                |
| :-- | :-- |
| Representación | Vector de valores reales |
| Recombinación  | No hay                   |
| Mutación       | Perturbación Gausiana    |
| Selección de Padres | Determinista: 1 padre muta a 1 hijo |
| Supervivencia  | Torneo Round-robin aletorio (1 vs 10) |
| Especialización | Auto-adaptación a ambientes vía mutación |

### Estrategias de Evolución (_ES, Evolution Strategies_)
* Rechenberg y Schwefel en 1960s, desarrollado en 1970s.
* Similar a EP con individuos de la misma especie $\rightarrow$ recombinación.
* Inicio el camino de Calibración de Parámetros en ejecución (on-line). 
* Posee dos estrategias ($\mu$, $\lambda$) y ($\mu + \lambda$)
    * ($\mu$, $\lambda$) $\rightarrow$ padres eliminados.
    * ($\mu + \lambda$) $\rightarrow$ padres e hijos disputan su avance.

| Caracteristica | Aplicación                |
| :-- | :-- |
| Representación | Vector de valores reales |
| Recombinación  | Discreta o Intermedia    |
| Mutación       | Perturbación Gausiana    |
| Selección de Padres | Aleatoria uniforme  |
| Supervivencia  | Elitismo |
| Especialización | Auto-adaptación a ambientes vía mutación |

### Algortimos Genéticos (_GA, Genetic Algorithms_)

* Holland (1973), rectificado por De Jong (1975).
* Algoritmo básico y pionero de la Computación Evolutiva.
* Creación de generaciones nuevas basadas en cruzamiento y mutación. 
* Población intermedia de padres que solo vive una (1) generación.


| Caracteristica | Apliación                |
| :-- | :-- |
| Representación | Bit-Vector or Bit-String |
| Recombinación  | Cruzamiento a 1 punto, probabilistico |
| Mutación       | Bit-flip (complemento de bit), probabilistico |
| Selección de Padres | Roulette-Wheel proporcional a Fitness  |
| Supervivencia  | Edad |

### Programación Genética (_GP, Genetic Programming_)

* J. Koza (1990), derivación de GA cambiando la representación a árboles como individuos.
* Dual: Optimización y Máquina de Aprendizaje.
    * Maximiza retorno.
    * Maximiza ajuste a modelo. 
* Moderno, pero tan existoso que tiene su propia rama.
* Uso: lenguajes, codigo, expresiones, logica, etc.

| Caracteristica | Apliación                |
| :-- | :-- |
| Representación | Árboles |
| Recombinación  | Intercambio de sub-árboles |
| Mutación       | Cambio aleatorio en el árbol. |
| Selección de Padres | Proporcional a Fitness (variadas técnicas)  |
| Supervivencia  | Edad |

## Derivaciones

### Co-Evolución

* Evolucion de una o varias poblaciones sometidas a diferentes tipos de evaluadores. 
* Si A es mejor que B, puede que en otro criterio B sea mejor que A. 
* Caso naturaleza:
    * Simbiosis: Co-evolución positiva.
    * Parasitismo: Co-evolución negativa.
<br><br>
* Mono-poblacional: Evaluadores son de la misma población. 
* Multi-poblacional: Evaluadores son de poblaciones diferentes.

### Algoritmos Bío-Inspirados
* Ant Colony Optimization: Búsqueda por población de hormigas.
    * ¿Como las hormigas encuentran alimento?
<br><br>
* Particle Swarm Optimization: Búsqueda por poblaciones lideradas por un miembro distinguible.
    * ¿Como viajan las bardadas de pájaros?
    * ¿Como se mueven los cárdumenes de peces?
