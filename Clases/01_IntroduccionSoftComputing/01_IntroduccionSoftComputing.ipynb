# Soft Computing

- El término **Soft Computing** fue introducido por Lotfi A. Zadeh.
- Soft Computing es un conjunto de técnicas capaces de resolver **problemas de alta complejidad** con técnicas **imprecisas o incompletas**.

# Contexto

En el mundo real, muchos problemas no tienen una solución exacta ni completamente determinista:
- Reconocer rostros o emociones.
- Controlar un robot en ambientes inciertos.
- Predecir el clima o la demanda eléctrica.

En estos casos, los métodos "estrictos" basados en reglas lógicas o algoritmos exactos fallan o resultan ineficientes. 
Softcomputing surge como una alternativa inspirada en la inteligencia humana: flexible, tolerante y adaptativa.

# Ejemplos

#### Estacionar automóvil en un espacio estrecho

No se necesitan las dimensiones **exactas** del espacio o del automóvil.

#### Reconocer caracteres escritos a mano

Aunque las letras escritas a mano no sigan su forma ideal, igualmente **somos capaces de reconocerlos**.

#### Hormigas recolectando comida

Inicialmente las hormigas buscan comida aleatoriamente. Cuando  una de ellas encuentra comida, **comparte el mensaje** con el resto para que todas se dirijan a la fuente de alimento.

# Definición de Soft Computing

Soft Computing combina múltiples técnicas para enfrentar problemas donde los métodos exactos son inaplicables o ineficientes. Sus principales características son:
- Tolerancia a la imprecisión.
- Capacidad de aprender y adaptarse.
- Soluciones aproximadas pero satisfactorias.

## Relación con la Inteligencia Artificial
Soft Computing es un subconjunto de técnicas de IA centrado en la heurística (conjunto de reglas o estrategias empíricas que guían la búsqueda de soluciones aproximadas en espacios complejos), el aprendizaje, y la optimización en contextos complejos.
A diferencia de los métodos clásicos de IA, que muchas veces dependen de reglas formales o modelos exactos, la softcomputación emplea:
- **Heurísticas**: estrategias prácticas y no necesariamente óptimas, que permiten encontrar soluciones razonables en espacios de búsqueda grandes o mal definidos.
- **Aprendizaje**: la capacidad de mejorar decisiones basadas en datos y experiencia.
- **Optimización flexible**: búsqueda de soluciones viables sin requerir exactitud absoluta.

En lugar de reemplazar los métodos deterministas, la softcomputación los complementa, ampliando el rango de problemas que pueden ser abordados con herramientas de IA. Su inspiración proviene tanto de la naturaleza como del comportamiento humano, donde la adaptabilidad y la tolerancia al error son esenciales.

# Componentes de Soft Computing

- **Lógica Difusa**: Permite razonar con grados de verdad, en vez de verdadero/falso. Ideal para sistemas de control.
- **Redes Neuronales Artificiales**: Sistemas que aprenden patrones desde datos. Utilizados en clasificación, regresión, y generación de contenido.
- **Algoritmos Evolutivos**: Inspirados en la selección natural, exploran soluciones mediante poblaciones y operadores como mutación y cruce.
- **Sistemas Híbridos**: Combinan dos o más técnicas anteriores, por ejemplo, redes neuronales con reglas difusas.

# Conceptos importantes

## Espacio (de Estados) de Búsqueda

- Corresponde al conjunto, discreto, de estados o configuraciones posibles para un sistema.
    * Cantidad de soluciones (factibles e infactibles) de un problema. 
    * Cantidad de posibles recomendaciones que una ANN puede entregar.
- Define la dificultad de un problema.
    * Conjunto finito de tamaño constante.
    * Conjunto finito de gran tamaño.
    * Conjunto finito creciente por instancia.
- Búsqueda: desde un **estado inicial** hacia un **estado final**.
    * Este estado final se define en base al problema a abordar o resolver.
    * Transformaciones que llevan desde un estadio inicial a uno final.

### Problema de Decisión

* Problema con resultado binario: ¿Sí o No?
    * ¿Se puede decidir si un número n es primo?
    * ¿Se puede decidir si los primeros n números naturales son primos?

* Buscamos: Clases de problemas.

### Máquina de Turing

* Modelo teórico propuesto por Alan Turing (1936) que define qué es un algoritmo y qué puede ser computado.
* Puede simular cualquier cálculo (modelo universal).

**Componentes**
- Cinta infinita (memoria).
- Cabezal lector-escritor (lee, escribe, se mueve).
- Conjunto finito de estados (inicial y de parada).
- Tabla de transición (reglas de acción según estado y símbolo).

**Funcionamiento**
- Inicia en un estado y celda.
- En cada paso: lee símbolo → aplica regla → escribe/mueve/cambia estado.
- Se detiene en un estado de parada o si no hay regla aplicable.

**Importancia**
- Fundamenta la teoría de la computación.
- Define qué problemas son computables y clasifica su complejidad (P, NP, etc.).


### Definición formal de la Máquina de Turing 

- Está compuesta de un alfabeto ($\Gamma$), input ($\Sigma$), estados ($Q$) y función de transición ($\delta$).  

$$
\delta : Q \times \Gamma \;\;\rightarrow\;\; \Gamma \times \{L, R\} \times Q
$$

donde $L$ (Left) y $R$ (Right) indican los movimientos de la cabeza lectora.

#### Ejemplo

$$
\Gamma = \{\square, 0, 1\} \quad \text{(Alfabeto)}
$$

$$
Q = \{q_0, q_1, q_2, q_3\} \quad \text{(Estados)}
$$

$$
\delta(q_0, 0) = \{0, R, q_1\} \quad \text{(Función de transición)}
$$



### Máquina de Turing No Determinista

* Extiende el modelo de Turing clásico permitiendo que, para una misma configuración (estado y símbolo leído), puedan existir **múltiples transiciones posibles**.
* En lugar de una única acción definida, la máquina puede "ramificarse" en varios caminos de cómputo al mismo tiempo.

**Definición formal**
- La función de transición devuelve un **conjunto de posibles acciones**:
  
$$
\delta : Q \times \Gamma \;\;\rightarrow\;\; \mathcal{P}(\Gamma \times \{L, R\} \times Q)
$$

donde $\mathcal{P}$ indica el conjunto de posibles opciones (poder elegir varias transiciones).

**Funcionamiento**
- En cada paso, si hay varias reglas aplicables, la máquina puede seguir cualquiera de ellas.
- El cómputo se entiende como un "árbol de ejecuciones".
- Se dice que la máquina **acepta** una entrada si al menos uno de los caminos de cómputo lleva a un estado de solución.

**Importancia**
- Es la base para la definición de la clase de complejidad **NP** (problemas verificables en tiempo polinómico).
- Aunque no existen máquinas no deterministas físicas, son un modelo teórico clave para entender qué problemas pueden resolverse eficientemente si se permitiera el no determinismo.

## Clase P

- Conjunto de problemas de decisión, que pueden ser resueltos por una **Máquina de Turing** *determinista* en tiempo polinomial.
  - Informalmente: Problemas que pueden ser resueltos (decididos) por un algoritmo determinista en tiempo polinomial.
  - Determinista: algoritmo que sigue siempre pasos conocidos.
  - Un algoritmo resuelve diferentes instancias de distintos problemas tipo P.
  - P = Polynomial.

## Clase NP

* Conjunto de problemas de decisión, que pueden ser resueltos por una **Maquina de Turing** _no determinista_ en tiempo polinomial.
    * Informalmente: Problemas decidibles (sí, por respuesta) cuyas soluciones pueden ser verificadas en tiempo polinomial.
    * Resolución: Difícil (a menos que también esté en P). 
    * No Determinista: algoritmo que no sigue siempre pasos conocidos.
    * Un algoritmo no resolverá de igual forma diferentes instancias de distintos problemas tipo NP.
    * NP = Non-deterministic Polynomial.

## NP-Complete

* **NP-Complete** son aquellos problemas de tipo *NP* tales que **cualquier otro** problema NP puede ser reducido a este en tiempo polinomial.

    * Reducción: función de transformación.

* **Siguen siendo NP**: Dificil de resolver, fácil de verificar solución.

* Papers base:
    * Todo SAT es NP-Completo. [1] Stephen Cook, The complexity of theorem-proving procedures. 1971. https://dl.acm.org/doi/10.1145/800157.805047
    * 21 Problemas NP-Completos [2] Richard Karp, Reducibility Among Combinatorial Problems, 1972. https://cgi.di.uoa.gr/~sgk/teaching/grad/handouts/karp.pdf

## NP-Hard

* Se les llama problemas **NP-Hard** a aquellos que, si bien **no necesariamente** se encuentran en el conjunto NP, pueden llegar a ser **tan difíciles** como los problemas más complejos de NP.
* Formalmente: Un problema es **NP-Hard** si un problema **NP-Completo** puede reducir a él.
    * Se consideran **Máquinas de Turing** _no deterministas_ para decidir (resolver) y verificar la solución al problema en tiempo polinomial.
* Un problema NP-Hard, puede incluso no ser un problema de decisión. 
* Informalmente: Problemas cuya resolución y verificación son difíciles. 
* Ejemplo más famoso: Travelling Salesman Problem (TSP).

## ¿P = NP?

* Problema esencial de la complejidad teórica.
* NP-Complete, NP-Hard: Importancia radical para resolver este problema.
    * **Si hubiese** una Maquina de Turing Determinista que resuelva en tiempo polinomial un *problema de decisión* NP-Hard, resolveremos todos los problemas NP-Complete.
    * **Si hubiese** una Maquina de Turing Determinista que resuelva en tiempo polinomial un problema NP-Complete, resolveremos todos los problemas tipo NP, incluyendo los de tipo P. 

* **Santo Grial**: Un algoritmo para resolver todo.

<img width="70%" src="images/p_np.svg" alt="" />

- El dilema **P vs NP** define qué problemas pueden resolverse de manera eficiente y cuáles requieren estrategias aproximadas.  
- Muchos problemas reales en sistemas difusos son **NP-hard**, lo que implica que no se conocen algoritmos polinomiales para resolverlos exactamente.  
- Soft Computing surge como respuesta: usa **heurísticas, metaheurísticas y técnicas aproximadas** para obtener soluciones útiles en tiempos razonables.  
- Entender P vs NP ayuda a justificar por qué se aplican métodos flexibles en lugar de algoritmos exactos.  
- Permite comunicar mejor el valor de los enfoques difusos: no buscan la perfección matemática, sino la **viabilidad práctica** en problemas complejos.
