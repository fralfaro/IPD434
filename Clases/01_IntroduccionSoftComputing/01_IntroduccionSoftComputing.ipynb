## 1. ¿Qué es el Soft Computing?

El **Soft Computing** (computación flexible o "suave") agrupa un conjunto de técnicas que buscan soluciones *aproximadas pero útiles* a problemas complejos, tolerando la imprecisión, la incertidumbre y la información incompleta. Se contrapone al **Hard Computing** (computación "dura"), que exige modelos exactos y datos precisos, y garantiza soluciones óptimas a costa de un mayor esfuerzo computacional.

- El término **Soft Computing** fue introducido por **Lotfi A. Zadeh** (1994), el mismo autor que propuso la lógica difusa en 1965.
- Es un conjunto de técnicas capaces de resolver **problemas de alta complejidad** a partir de información **imprecisa, incierta o incompleta**.
- Su principio rector es la **tolerancia a la imprecisión**: aceptar cierto margen de error a cambio de robustez, bajo costo y tratabilidad.

## 2. Contexto: por qué necesitamos Soft Computing

En el mundo real, muchos problemas no tienen una solución exacta ni completamente determinista:
- Reconocer rostros o emociones.
- Controlar un robot en ambientes inciertos.
- Predecir el clima o la demanda eléctrica.

En estos casos, los métodos "estrictos" basados en reglas lógicas o algoritmos exactos fallan o resultan ineficientes. El Soft Computing surge como una alternativa inspirada en la inteligencia humana: flexible, tolerante y adaptativa.

## 3. Ejemplos motivadores

Antes de formalizar, revisemos tres situaciones cotidianas en las que resolvemos problemas *sin* información exacta ni un algoritmo óptimo. Cada una anticipa un paradigma del Soft Computing.

### Estacionar un automóvil en un espacio estrecho

No se necesitan las dimensiones **exactas** del espacio o del automóvil; basta con reglas cualitativas ("gira un poco más", "avanza lento"). → Anticipa la **Lógica Difusa**.

### Reconocer caracteres escritos a mano

Aunque las letras escritas a mano no sigan su forma ideal, igualmente **somos capaces de reconocerlas** a partir de ejemplos vistos antes. → Anticipa las **Redes Neuronales**.

### Hormigas recolectando comida

Inicialmente las hormigas buscan comida de forma aleatoria. Cuando una de ellas la encuentra, **comparte el mensaje** (feromonas) con el resto para que todas converjan hacia la fuente de alimento. → Anticipa la **Computación Evolutiva / Inteligencia de Enjambre**.

## 4. Incertidumbre, imprecisión y vaguedad

El Soft Computing existe porque el mundo real rara vez nos entrega información perfecta. Conviene distinguir tres fuentes de "información imperfecta" que suelen confundirse, porque cada una motiva herramientas distintas.

### Conceptos clave

- **Incertidumbre (uncertainty):** no sabemos con certeza si un hecho es verdadero o cuál será el resultado; podemos, a lo sumo, asignarle una *probabilidad*. *Ejemplo:* "mañana lloverá con probabilidad 0.7".
- **Imprecisión (imprecision):** la medición o el dato tiene un margen de error o una resolución limitada. *Ejemplo:* "la temperatura es 20 °C ± 1 °C".
- **Vaguedad (vagueness):** el concepto en sí no tiene una frontera nítida; admite *grados* de pertenencia. *Ejemplo:* "el agua está caliente" — no hay un umbral exacto que separe "caliente" de "no caliente".

### Sistemas deterministas vs. sistemas con incertidumbre

- **Sistema determinista:** dada la misma entrada, produce siempre la misma salida y su comportamiento se describe con reglas o ecuaciones exactas (p. ej. ordenar una lista).
- **Sistema con incertidumbre:** la entrada, el modelo o la salida son parcialmente desconocidos, ruidosos o vagos, por lo que se busca una respuesta *robusta* más que exacta (p. ej. un diagnóstico médico a partir de síntomas).

| Fuente | Pregunta que responde | Herramienta típica |
|--------|-----------------------|--------------------|
| Incertidumbre | ¿Qué tan probable es? | Probabilidad / Bayes |
| Imprecisión | ¿Con qué margen de error? | Intervalos, estadística |
| Vaguedad | ¿En qué grado pertenece? | **Lógica Difusa** |

> 💡 La **vaguedad** es precisamente lo que la lógica difusa modela con *grados de verdad* entre 0 y 1, a diferencia de la probabilidad, que cuantifica la incertidumbre.

## 5. Definición de Soft Computing

Soft Computing combina múltiples técnicas para enfrentar problemas donde los métodos exactos son inaplicables o ineficientes. Sus principales características son:
- Tolerancia a la imprecisión.
- Capacidad de aprender y adaptarse.
- Soluciones aproximadas pero satisfactorias.

### Relación con la Inteligencia Artificial

Soft Computing es un subconjunto de técnicas de IA centrado en la **heurística**, el aprendizaje y la optimización en contextos complejos. Una **heurística** es un conjunto de reglas o estrategias empíricas que guían la búsqueda de soluciones aproximadas en espacios complejos.

A diferencia de los métodos clásicos de IA, que muchas veces dependen de reglas formales o modelos exactos, el Soft Computing emplea:
- **Heurísticas:** estrategias prácticas y no necesariamente óptimas, que permiten encontrar soluciones razonables en espacios de búsqueda grandes o mal definidos.
- **Aprendizaje:** la capacidad de mejorar decisiones basadas en datos y experiencia.
- **Optimización flexible:** búsqueda de soluciones viables sin requerir exactitud absoluta.

En lugar de reemplazar los métodos deterministas, el Soft Computing los complementa, ampliando el rango de problemas que pueden ser abordados con herramientas de IA. Su inspiración proviene tanto de la naturaleza como del comportamiento humano, donde la adaptabilidad y la tolerancia al error son esenciales.

## 6. Componentes del Soft Computing

El Soft Computing no es una sola técnica, sino una **familia** de paradigmas complementarios. Los tres pilares clásicos son la lógica difusa, las redes neuronales y la computación evolutiva; su combinación da lugar a los sistemas híbridos.

- **Lógica Difusa (Fuzzy Logic):** permite razonar con *grados de verdad* en vez de verdadero/falso absoluto. Ideal para sistemas de control basados en conocimiento experto.
- **Redes Neuronales Artificiales (ANN):** sistemas que **aprenden patrones desde datos**. Se usan en clasificación, regresión y generación de contenido.
- **Algoritmos Evolutivos (EA):** inspirados en la selección natural, exploran el espacio de soluciones mediante *poblaciones* y operadores como mutación y cruce. Útiles en optimización difícil.
- **Sistemas Híbridos:** combinan dos o más técnicas anteriores; por ejemplo, redes neuronales con reglas difusas (*neuro-fuzzy*) o EA que ajustan los pesos de una red.

## 7. ¿Cómo elegir el enfoque adecuado?

No existe un paradigma "mejor" en todos los casos: la elección depende de la naturaleza del problema, de los datos disponibles y de lo que se quiera optimizar. La siguiente tabla resume criterios prácticos para **clasificar un problema** y decidir por dónde empezar.

### Criterios de selección

| Si el problema... | Enfoque sugerido |
|-------------------|------------------|
| Se describe con reglas expertas del tipo *SI... ENTONCES* y conceptos vagos | **Lógica Difusa** |
| Dispone de muchos datos etiquetados y busca reconocer patrones | **Redes Neuronales** |
| Requiere optimizar/buscar en un espacio enorme sin gradiente claro | **Computación Evolutiva** |
| Mezcla conocimiento experto + datos + optimización | **Sistema Híbrido** |

### Caso motivador: control de temperatura de una sala

Consideremos mantener una sala a una temperatura confortable. El mismo problema admite varias soluciones de Soft Computing:

- **Difuso:** reglas como *"SI hace frío ENTONCES calefacción alta"*, sin necesidad de un modelo físico exacto.
- **Neuronal:** aprender de datos históricos (temperatura, hora, ocupación) la potencia óptima del climatizador.
- **Evolutivo:** buscar los parámetros del controlador que minimizan el consumo energético sujeto al confort.

### Métricas para comparar enfoques

Al enfrentar un mismo problema con distintos paradigmas, conviene compararlos con criterios explícitos:

| Métrica | Qué mide |
|---------|----------|
| **Precisión / error** | Qué tan buena es la solución (accuracy, RMSE, confort). |
| **Costo computacional** | Tiempo y memoria para obtener la solución. |
| **Interpretabilidad** | Cuán fácil es explicar por qué el sistema decide así. |
| **Datos requeridos** | Cantidad de datos etiquetados necesarios. |
| **Robustez** | Tolerancia al ruido y a entradas imprecisas. |

## 8. Conceptos importantes

Para entender *por qué* el Soft Computing renuncia a la solución óptima exacta, necesitamos dos ideas de las Ciencias de la Computación: el **espacio de búsqueda** (cuántas soluciones posibles existen) y la **complejidad computacional** (cuán costoso es explorarlo).

### 8.1 Espacio de estados de búsqueda

El **espacio de búsqueda** es el conjunto de todas las configuraciones (estados) posibles que un problema puede tomar. Su tamaño determina qué tan difícil es encontrar una buena solución.

- Corresponde al conjunto, discreto, de estados o configuraciones posibles para un sistema.
    * Cantidad de soluciones (factibles e infactibles) de un problema.
    * Cantidad de posibles recomendaciones que una ANN puede entregar.
- Define la dificultad de un problema.
    * Conjunto finito de tamaño constante.
    * Conjunto finito de gran tamaño.
    * Conjunto finito creciente por instancia.
- Búsqueda: desde un **estado inicial** hacia un **estado final**.
    * Este estado final se define en base al problema a abordar o resolver.
    * Transformaciones que llevan desde un estado inicial a uno final.

#### Ejemplo: el juego del "tres en raya" (*tic-tac-toe*)

Cada tablero es un **estado**; partiendo del tablero vacío, cada jugada genera nuevos estados hasta llegar a un estado final (victoria, derrota o empate). El árbol siguiente ilustra parte de ese espacio de estados.

<center><img src="images/tictactoe.png" alt="Árbol de estados del juego tres en raya" width="75%" align="center"/></center>

### 8.2 Tipos de búsqueda

Según cuánto del espacio exploremos, distinguimos búsqueda **completa** (garantiza el óptimo) e **incompleta** (renuncia a la garantía a cambio de eficiencia).

* **Búsqueda completa:** examina *todo* el espacio de estados para hallar el óptimo **global**.
    * Corresponde al **Hard Computing** (garantiza la mejor solución, pero puede ser inviable en espacios grandes).
* **Búsqueda incompleta:** examina solo una parte del espacio y devuelve un óptimo **local** o una buena aproximación.
    * Corresponde al **Soft Computing** (renuncia a la garantía de optimalidad a cambio de tratabilidad).

## 9. Complejidad computacional de problemas

La **complejidad computacional** clasifica los problemas según los recursos (tiempo, memoria) que requiere resolverlos. Muchos problemas de interés son tan costosos que su resolución exacta es inviable, lo que **justifica** el uso de las técnicas aproximadas del Soft Computing.

### 9.1 Problema de decisión

Un **problema de decisión** es aquel cuya respuesta es binaria (**sí** o **no**). La teoría de la complejidad se construye sobre este tipo de problemas porque permiten agruparlos en *clases*.

* Ejemplos:
    * ¿Es primo un número $n$?
    * ¿Son primos los primeros $n$ números naturales?
* Objetivo: agrupar los problemas en **clases de complejidad**.

### 9.2 Máquina de Turing

La **Máquina de Turing** es el modelo teórico que formaliza la noción de *algoritmo* y sirve de base para clasificar la complejidad de los problemas.

* Modelo teórico propuesto por Alan Turing (1936) que define qué es un algoritmo y qué puede ser computado.
* Puede simular cualquier cálculo (modelo universal).

**Componentes**
- Cinta infinita (memoria).
- Cabezal lector-escritor (lee, escribe, se mueve).
- Conjunto finito de estados (inicial y de parada).
- Tabla de transición (reglas de acción según estado y símbolo).

**Funcionamiento**
- Inicia en un estado y celda.
- En cada paso: lee símbolo → aplica regla → escribe/mueve/cambia estado.
- Se detiene en un estado de parada o si no hay regla aplicable.

**Importancia**
- Fundamenta la teoría de la computación.
- Define qué problemas son computables y clasifica su complejidad (P, NP, etc.).

#### Definición formal

- Está compuesta de un alfabeto ($\Gamma$), input ($\Sigma$), estados ($Q$) y función de transición ($\delta$).

$$
\delta : Q \times \Gamma \;\;\rightarrow\;\; \Gamma \times \{L, R\} \times Q
$$

donde $L$ (*Left*) y $R$ (*Right*) indican los movimientos de la cabeza lectora.

<center><img src="images/cinta_turing.png" alt="Esquema de la cinta de una Máquina de Turing" width="35%" align="center"/></center>

##### Ejemplo

$$
\Gamma = \{\square, 0, 1\} \quad \text{(Alfabeto)}
$$

$$
Q = \{q_0, q_1, q_2, q_3\} \quad \text{(Estados)}
$$

$$
\delta(q_0, 0) = \{0, R, q_1\} \quad \text{(Función de transición)}
$$

### 9.3 Máquina de Turing no determinista

* Extiende el modelo de Turing clásico permitiendo que, para una misma configuración (estado y símbolo leído), puedan existir **múltiples transiciones posibles**.
* En lugar de una única acción definida, la máquina puede "ramificarse" en varios caminos de cómputo al mismo tiempo.

**Definición formal**
- La función de transición devuelve un **conjunto de posibles acciones**:

$$
\delta : Q \times \Gamma \;\;\rightarrow\;\; \mathcal{P}(\Gamma \times \{L, R\} \times Q)
$$

donde $\mathcal{P}$ indica el conjunto de posibles opciones (poder elegir varias transiciones).

**Funcionamiento**
- En cada paso, si hay varias reglas aplicables, la máquina puede seguir cualquiera de ellas.
- El cómputo se entiende como un "árbol de ejecuciones".
- Se dice que la máquina **acepta** una entrada si al menos uno de los caminos de cómputo lleva a un estado de solución.

**Importancia**
- Es la base para la definición de la clase de complejidad **NP** (problemas verificables en tiempo polinómico).
- Aunque no existen máquinas no deterministas físicas, son un modelo teórico clave para entender qué problemas pueden resolverse eficientemente si se permitiera el no determinismo.

### 9.4 Clase P

- Conjunto de problemas de decisión que pueden ser resueltos por una **Máquina de Turing** *determinista* en tiempo polinomial.
  - Informalmente: problemas que pueden ser resueltos (decididos) por un algoritmo determinista en tiempo polinomial.
  - Determinista: algoritmo que sigue siempre pasos conocidos.
  - Un mismo algoritmo resuelve diferentes instancias de un problema tipo P.
  - P = *Polynomial*.

### 9.5 Clase NP

* Conjunto de problemas de decisión que pueden ser resueltos por una **Máquina de Turing** *no determinista* en tiempo polinomial.
    * Informalmente: problemas decidibles (respuesta *sí*) cuyas soluciones pueden ser **verificadas** en tiempo polinomial.
    * Resolución: difícil (a menos que también estén en P).
    * No determinista: algoritmo que no sigue siempre pasos conocidos.
    * Un algoritmo no resolverá de igual forma diferentes instancias de distintos problemas tipo NP.
    * NP = *Non-deterministic Polynomial*.

### 9.6 NP-Completo

* Los problemas **NP-Completos** son aquellos problemas de tipo *NP* tales que **cualquier otro** problema NP puede ser reducido a ellos en tiempo polinomial.

    * Reducción: función de transformación entre problemas.

* **Siguen siendo NP:** difícil de resolver, fácil de verificar la solución.

* Papers base:
    * Todo SAT es NP-Completo. [1] Stephen Cook, *The complexity of theorem-proving procedures*, 1971. https://dl.acm.org/doi/10.1145/800157.805047
    * 21 problemas NP-Completos. [2] Richard Karp, *Reducibility Among Combinatorial Problems*, 1972. https://cgi.di.uoa.gr/~sgk/teaching/grad/handouts/karp.pdf

### 9.7 NP-Hard

* Se les llama problemas **NP-Hard** a aquellos que, si bien **no necesariamente** se encuentran en el conjunto NP, pueden llegar a ser **tan difíciles** como los problemas más complejos de NP.
* Formalmente: un problema es **NP-Hard** si un problema **NP-Completo** puede reducirse a él.
    * Se consideran **Máquinas de Turing** *no deterministas* para decidir (resolver) y verificar la solución al problema en tiempo polinomial.
* Un problema NP-Hard puede incluso no ser un problema de decisión.
* Informalmente: problemas cuya resolución y verificación son difíciles.
* Ejemplo más famoso: el **Problema del Vendedor Viajero** (*Travelling Salesman Problem*, TSP).

### 9.8 ¿P = NP?

* Problema esencial de la complejidad teórica.
* NP-Completo y NP-Hard tienen una importancia radical para resolver este problema:
    * **Si hubiese** una Máquina de Turing determinista que resuelva en tiempo polinomial un *problema de decisión* NP-Hard, resolveríamos todos los problemas NP-Completos.
    * **Si hubiese** una Máquina de Turing determinista que resuelva en tiempo polinomial un problema NP-Completo, resolveríamos todos los problemas tipo NP, incluyendo los de tipo P.

* **Santo Grial:** un algoritmo eficiente para resolverlo todo.

<center><img width="70%" src="images/p_np.svg" alt="Diagrama de relaciones entre las clases P, NP, NP-Completo y NP-Hard" /></center>

- El dilema **P vs NP** define qué problemas pueden resolverse de manera eficiente y cuáles requieren estrategias aproximadas.
- Muchos problemas reales de optimización son **NP-Hard**, lo que implica que no se conocen algoritmos polinomiales para resolverlos exactamente.
- El Soft Computing surge como respuesta: usa **heurísticas, metaheurísticas y técnicas aproximadas** para obtener soluciones útiles en tiempos razonables. Una **metaheurística** es una estrategia general de búsqueda (p. ej. algoritmos evolutivos) aplicable a muchos problemas distintos.
- Entender P vs NP ayuda a justificar por qué se aplican métodos flexibles en lugar de algoritmos exactos.
- Permite comunicar mejor el valor de estos enfoques: no buscan la perfección matemática, sino la **viabilidad práctica** en problemas complejos.

* [3] Richard E. Ladner, *On the Structure of Polynomial Time Reducibility*, 1975. https://dl.acm.org/doi/10.1145/321864.321877

## 📌 Resumen

| Concepto | Idea clave |
|----------|-----------|
| **Soft Computing** | Familia de técnicas que toleran imprecisión, incertidumbre y datos incompletos para hallar soluciones *aproximadas pero útiles*. |
| **Hard vs. Soft** | Hard Computing busca el óptimo exacto (búsqueda completa); Soft Computing acepta óptimos locales (búsqueda incompleta). |
| **Incertidumbre / imprecisión / vaguedad** | Tres formas de información imperfecta; la vaguedad es el dominio propio de la lógica difusa. |
| **Componentes** | Lógica Difusa, Redes Neuronales, Computación Evolutiva y sus híbridos. |
| **Elección del enfoque** | Depende de los datos, del tipo de conocimiento disponible y de la métrica a optimizar. |
| **Complejidad (P, NP, NP-Hard)** | Los problemas intratables exactamente (NP-Hard, como el TSP) justifican el uso de heurísticas. |
